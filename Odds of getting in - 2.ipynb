### Make separate dataframes for applications and offers

```python
apps = df[df["file"] == "apps"]
offers = df[df["file"] == "offers"]
print len(apps) + len(offers)
```

```python
apps.head()
```

### How many schools are in apps but not offers

```python
def in_a_not_b(a, b, col="school"):
    return a[a[col].apply(lambda x: x not in b[col].unique())][col].unique()
```

```python
in_a_not_b(apps,offers)
```

```python
in_a_not_b(offers,apps)
```

### Combine the data

```python
ind=["operator","school","grade"]
```

```python
combined = apps.set_index(ind).join(offers.set_index(ind), how="outer",rsuffix="_offers")
combined.reset_index()
```

```python
full_ind = ["operator",
            "school",
            "grade",
            "n_apps",
            "n_apps_hartford",
            "n_apps_suburban",
            "n_offers_offers",
            "n_offers_hartford_offers",
            "n_offers_suburban_offers"
           ]
```

```python
combined_pared_down = combined.reset_index()[full_ind]
combined_pared_down
```

```python
combined_pared_down.columns = ["operator","school","grade","n_apps","n_apps_hartford","n_apps_suburban",
                               "n_offers","n_offers_hartford","n_offers_suburban"]
combined_pared_down.sort_values(by=["operator","school","grade"])
```

```python
combined_pared_down.sort_values(by=["operator","school","grade"]).to_csv("data/hopefully-clean-odds.csv",index=False)
```
