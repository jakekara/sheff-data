```python
new_data = deseg_15[["Unnamed: 1", "% Black/ Latino", "% Reduced- Isolation"]].fillna(-1)
new_data.columns = "School","Black_Latino_Pct","Reduced_Isolation_Pct"
new_data = new_data[new_data["School"]!=-1]
new_data["School"] = new_data.apply(lambda x: x["School"].replace("*",""), axis=1)
new_data
```

```python
names = pd.read_csv("data/name_replace.csv")
names["School_16"] = names["School_16"].str.replace("*","")
names["School_16"] = names["School_16"].str.strip()

names
```

```python
def get_lat(school_name):
    
    print "'" + school_name + "'"
    df = names[names["School_16"].replace(",","") == school_name.strip()]
    #print df
    if len(df.index) < 1:
        ret = -1,-1
    else:
        ret = float(df["lat"].replace(",","")), float(df["lng"])
    print ret
    return ret

new_data["lat"] = new_data.apply(lambda x: get_lat(x["School"])[0],axis=1)
new_data["lng"] = new_data.apply(lambda x: get_lat(x["School"])[1],axis=1)
```

```python
new_data["Black_Latino_Pct"] = new_data["Black_Latino_Pct"].round()
new_data["Reduced_Isolation_Pct"] = new_data["Reduced_Isolation_Pct"].round()

new_data
```
