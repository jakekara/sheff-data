```python
import pandas as pd
from ct_schools import schools
```

```python
wait_list = pd.read_excel("data/Ct Mirror FOI 1-19 - Wait List.xlsx")
wait_list.head()
```

```python
# Let's clean these column names
for c in wait_list.columns: print c
```

```python
wait_list.columns = ["operator",
                     "school_first_choice",
                     "wait_list_hartford_bl",
                     "wait_list_hartford_ri",
                     "wait_list_sub_all",
                     "wait_list_sub_ri",
                     "first_10_ri",
                     "first_20_ri"
                    ]
wait_list.head()
```

```python
wait_list = wait_list.dropna(subset=["school_first_choice"])
```

```python
wait_list
```

```python
wait_list.dtypes
```

```python
# Get the best guesses for school names
# and then list them in order of lowest fuzzy ratio (worst match)
# to highest ratio (best match) so we can individually examine
# some of the more questionable matches

wait_list["name_guess"] = wait_list["school_first_choice"].apply(lambda x: schools.school_by_name(x, closest=True).iloc[0]["School Name"])
wait_list["fuzz_ratio"] = wait_list["school_first_choice"].apply(lambda x: schools.school_by_name(x, closest=True).iloc[0]["FUZZ_RATIO"])
wait_list["lat"] = wait_list["school_first_choice"].apply(lambda x: schools.school_by_name(x, closest=True).iloc[0]["lat"])
wait_list["lng"] = wait_list["school_first_choice"].apply(lambda x: schools.school_by_name(x, closest=True).iloc[0]["lng"])
wait_list["lat_lng"] = wait_list["school_first_choice"].apply(lambda x: schools.school_by_name(x, closest=True).iloc[0]["lat_lng"])


```

```python
wait_list[["school_first_choice","name_guess","fuzz_ratio"]].sort_values(by="fuzz_ratio")
```
